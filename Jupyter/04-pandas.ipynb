#### **Eliminar datos nulos del dataset**

Importante = Guardar el dataset sin nulos dentro de una variable

```python
datos_sin_nulos = datos.dropna()
datos_sin_nulos
```

```python
#Funcion que filtra por genero RPG
def filtro_rpg(genres):
    return"RPG" in genres

#Aplicar un filtro para consultar y llamar a la funcion filtro_rpg
datosRPG = datos_sin_nulos[datos_sin_nulos["Genres"].apply(filtro_rpg)]


nombresRPG = datosRPG[["Title", "Rating"]]
nombresRPG.head(20)
```
