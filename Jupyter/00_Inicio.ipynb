#### **Montando Unidad de Google Drive (No ejecutar en VS Code)**
Este bloque de código es solo para usuarios que utilicen Google Colaboratory como entorno de desarrollo, no servirá para un kernel local de VS Code/Jupyter. Lo que hace es montar (conectar) tu cuenta de Google Drive en el entorno virtual de Google Colab. Permite leer, editar y guardar archivos directamente desde tus carpetas de Drive utilizando Python.

###### *Nota: Por seguridad, deberás autorizar este acceso cada vez que inicies un nuevo entorno de ejecución en Colab.*

Se agrega la ruta donde se aloja el dataset en Google Drive. En este caso la variable `ruta` representa la ubicación del archivo que se desea abrir.

```python
ruta = "/content/drive/MyDrive/Colab/data/games.csv"
```

En esta línea se carga un archivo de formato CSV que es un DataFrame de pandas. La función `pd.read_csv()` lee lo datos tabulares desde un archivo y los convierte en una estructura para que Python lo manipule.

El parámetro `index_col='Id'` indica que la columna llamada `Id` se usará como índice de las filas del DataFrame resultante, que quedará almacenado en la variable datos.



```python
datos = pd.read_csv(ruta, index_col='Id')
```

A continuación el método `.head()` muestra las primeras filas del DataFrame `datos`. Es una forma rápida de inspeccionar el contenido de un dataset sin tener que mostrar todo el archivo.

Por defecto `.head()` devuelve las primeras 5 filas, aunque se puede indicar un número distinto.

```python
datos.head()
```
