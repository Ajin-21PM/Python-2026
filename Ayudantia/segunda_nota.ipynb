### **Importando librerias**

```python
import numpy as np
import matplotlib.pyplot as plt
```

### **lenguajes y preferencias**

```python
lenguajes = ["Python", "C++", "Java", "JavaScript", "C#"]
print(lenguajes)
preferencia = [45, 20, 15, 12, 8]
print(preferencia)
```

### **Grafico de Barra**

```python

plt.figure(figsize=(4,4))
plt.bar(lenguajes, preferencia, color = ["red","black", "orange", "blue", "yellow"])
plt.title("Lenguajes de programacion", fontsize=12, fontweight="bold")
plt.xlabel("lenguajes", fontsize= 8)
plt.ylabel("preferencia", fontsize= 8) 
plt.show 
```
